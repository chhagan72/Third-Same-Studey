```python
import matplotlib.pyplot as plt
# scipy is scientific computation library
from scipy import stats
x=[5,7,8,7,2,17,2,9,4,11,12,9,6,5,7]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86,66,79]
# y=ax+b --> a is slope and b is intercept

slope,intercept,r,p,std_err=stats.linregress(x,y)
def my_func(x):
    return slope*x+intercept # create a func that uses slope and intercept value to return a new vlaue
myModel=list(map(my_func,x)) # each value of x array and result stored in list it will map
plt.scatter(x,y) # original scatter graph
plt.plot(x,myModel) # x and mapping result will together draw graph
plt.show()
```

```python
my_func(10)
```
